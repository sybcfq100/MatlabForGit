```python
import numpy as np
```

```python
Y, y, R, r, g, t, sigma, delta = sympy.symbols("Y, y, R, r, g, t, sigma, delta")
```

```python
Y = sympy.Function("Y")(t)
y = sympy.Function("y")(t)
dY = sympy.diff(Y,t)
dy = sympy.diff(y,t)
```

```python
expr1 = sigma*r*r*sympy.sqrt(2*g*(Y-y)+dY*dY-dy*dy)/(R-delta)/(R-delta)
expr1
```

```python
expr2 = sigma*r*r*sympy.sqrt(2*g*(Y-y)+dY*dY-dy*dy)/R/R
expr2
```

```python
sympy.solve([expr1, expr2], [Y,y])
```
