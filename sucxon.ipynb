```python
Y, y, R, r, g, t, sigma, delta = sympy.symbols("Y, y, R, r, g, t, sigma, delta")
```

```python
Y = sympy.Function("Y")(t)
y = sympy.Function("y")(t)
dY = sympy.diff(Y,t)
dy = sympy.diff(y,t)
```

```python
expr1 = sigma*r*r*sympy.sqrt(2*g*(Y-y)+dY*dY-dy*dy)/(R-delta)/(R-delta)-dy
wt = np.array(50)
# t = sympy.symbols('wt')
values = {R:10, r:1, g:9.8, sigma:0.15, delta:0.2, t:wt}
Ve1 = expr1.subs(values)
Ve1
```

```python
expr2 = sigma*r*r*sympy.sqrt(2*g*(Y-y)+dY*dY-dy*dy)/R/R-dY
Ve2 = expr2.subs(values)
Ve2
```

```python
# sols = sympy.solve([Ve1, Ve2], [Y, y])
def funcs(X):
    y, Y = X
    f = [Ve1, Ve2]
    return f

sol2 = root(funcs, [0, 0])
```
